```python
import pandas as pd

#CODE TO EXTRACT SELECTED COLUMN FROM THE BASE MASTER DATA OF DIVISION
all_data = []

# Process each CSV file in the input_files folder
for file_name in os.listdir('input_files'):
    if file_name.endswith('.csv'):
        file_path = os.path.join('input_files', file_name)

        print(f"Processing file: {file_path}")

        # Read the CSV
        df = pd.read_csv(file_path)

        # Extract specific columns (change these columns as per your CSV)
        extracted = df[['ACCT_ID','DIV_CODE','SDO_CODE','TOWN', 'MOBILE_NO', 'SUPPLY_TYPE','CON_STATUS','CONNECTION_TYPE', 'BILL_CYC_CD','SERIAL_NBR', 'METER_BADGE_NO', 'MTR_TYPE_CD', 'METER_STATUS', 'METER_READ_REMARK', 'LAST_BILL_DATE', 'BILL_BASIS', 'MDI']]

        # Save each extracted file to Extracted_files folder
        extracted_file_path = os.path.join('Extracted_files', f"extracted_{file_name}")
        extracted.to_csv(extracted_file_path, index=False)

        print(f"Extracted data saved to: {extracted_file_path}")

        # Add to the list for final merge
        all_data.append(extracted)

# Combine all extracted data into one DataFrame
final_output = pd.concat(all_data, ignore_index=True)

# Save the combined data to All_Files_Data folder
final_output_path = 'All_Files_Data/final_output.csv'
final_output.to_csv(final_output_path, index=False)

print(f"✅ Final combined data saved to: {final_output_path}")
```

```python
import pandas as pd
import sqlite3


#SQL QUERY CODE FOR TWO FILES
file_path1 = os.path.join('input_files', "bill cycle file.csv")
file_path2 = os.path.join('input_files', "BI Monthly JULY 25.xlsx")

# Read Excel data
df1 = pd.read_csv(file_path1)
df2 = pd.read_excel(file_path2)

# Create SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Load DataFrames into SQL tables
df1.to_sql("table1", conn, index=False, if_exists="replace")
df2.to_sql("table2", conn, index=False, if_exists="replace")

# Run SQL JOIN
query = """
SELECT t1.*
FROM table1 t1, table2 t2
Where t1.BILL_CYC_CD = 'MD03' and t1.ACCT_ID=t2.ACCT_ID
"""
df_joined = pd.read_sql_query(query, conn)
conn.close()

# Save the combined data to All_Files_Data folder
final_output_path = 'All_Files_Data/df_joined.csv'
df_joined.to_csv(final_output_path, index=False)

print(f"✅ Final combined data saved to: {final_output_path}")
```
